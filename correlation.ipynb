```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
data = pd.read_csv('data.csv')
```

```python
plt.scatter(data['sum'], data['converted gpa'])
plt.xlabel('Social Media Usage')
plt.ylabel('GPA')
plt.show()
```

```python
# select columns for social media use and GPA
social_media = data['sum']
gpa = data['converted gpa']

# calculate correlation coefficient
corr_coef, p_val = pearsonr(social_media, gpa)

# print the correlation matrix
print("Pearson correlation coefficient: {:.2f}".format(corr_coef))
print("p-value: {:.4f}".format(p_val))


print("\n")
print("---------------------------------------------")
print("Since the pearson coefficient is negative and the p-value comes out to be statistically significant (less than 0.05), \nwe can conclude that there is a significant negative correlation between the two variables, i.e., use of social media and GPA.\n")
print("This means that higher the use of social media, lesser the GPA, thus proving our hypothesis.")
```
